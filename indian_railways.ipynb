```python
result_list = []

for (train_no, train_name), group in df.groupby(["Train No", "Train Name"]):
    stations = group["Station Name"].tolist()
    result_list.append({
        "train no": train_no,
        "train name": train_name,
        "stations": stations
    })
```

```python
data = pd.DataFrame.from_dict(result_list)
data.tail(10)
```

```python
import pandas as pd
import networkx as nx
from pyvis.network import Network
from collections import Counter

# -------------------------------
# 2. Count station frequencies
# -------------------------------
station_counter = Counter()
for stations_str in data['stations']:
    stations = stations_str
    station_counter.update(stations)

# -------------------------------
# 3. Identify major and hub stations
# -------------------------------
major_threshold = 3  # Appears in >=3 trains
major_stations = {st for st, count in station_counter.items() if count >= major_threshold}

# Identify top 1% busiest hubs
top_n = max(1, int(0.01 * len(station_counter)))  
hub_stations = set([st for st, _ in station_counter.most_common(top_n)])

# -------------------------------
# 4. Build simplified graph
# -------------------------------
G = nx.Graph()

for _, row in data.iterrows():
    stations = row['stations']
    
    simplified = []
    for st in stations:
        if st in major_stations:
            simplified.append(st)
        else:
            simplified.append(f"Cluster_{st[0]}")  # cluster by first letter

    for i in range(len(simplified)-1):
        if simplified[i] != simplified[i+1]:
            G.add_edge(simplified[i], simplified[i+1])

print(f"Original stations: {len(station_counter)}")
print(f"Major stations: {len(major_stations)}")
print(f"Hub stations: {len(hub_stations)}")
print(f"Simplified graph: {len(G.nodes())} nodes, {len(G.edges())} edges")

# -------------------------------
# 5. Create PyVis Graph with Colors
# -------------------------------
net = Network(
    height="900px", 
    width="100%", 
    bgcolor="#111111",  # Dark background
    font_color="white", 
    notebook=False
)
net.force_atlas_2based(gravity=-50)

for node in G.nodes():
    # Decide color based on type
    if node.startswith("Cluster_"):
        color = "#FFA500"  # Orange for clusters
        size = 5
    elif node in hub_stations:
        color = "#FF4500"  # Red for top hubs
        size = 20
    else:
        color = "#00BFFF"  # Blue for major stations
        size = 10
    
    net.add_node(node, label=node, color=color, size=size)

# Add edges
for u, v in G.edges():
    net.add_edge(u, v, color="#888888")

# -------------------------------
# 6. Save Interactive Graph
# -------------------------------
output_file = "train_network_colored.html"
net.save_graph(output_file)
print(f"Colored interactive graph saved as {output_file}. Open it in your browser.")
```

```python
import pandas as pd
import networkx as nx
from pyvis.network import Network
from collections import Counter

# -------------------------------
# 2. Define train color function
# -------------------------------
def get_train_color(train_name):
    name = str(train_name).upper()
    if "EMU" in name:
        return "#00FF00"  # Green
    elif "EXP" in name or "EXPRESS" in name:
        return "#0000FF"  # Blue
    elif "MAIL" in name or "SUPER" in name:
        return "#FF0000"  # Red
    else:
        return "#AAAAAA"  # Gray for others

# -------------------------------
# 3. Count station frequencies
# -------------------------------
station_counter = Counter()
for stations in data['stations']:
    station_counter.update(stations)

major_threshold = 3  # Appears in >=3 trains
major_stations = {st for st, count in station_counter.items() if count >= major_threshold}

# Identify top 1% busiest hubs
top_n = max(1, int(0.01 * len(station_counter)))  
hub_stations = set([st for st, _ in station_counter.most_common(top_n)])

print(f"Original stations: {len(station_counter)}")
print(f"Major stations (>=3 trains): {len(major_stations)}")
print(f"Hub stations (top 1% busiest): {len(hub_stations)}")

# -------------------------------
# 4. Build simplified graph
# -------------------------------
G = nx.Graph()

for _, row in data.iterrows():
    stations = row['stations']
    train_color = get_train_color(row['train name'])

    # Simplify station names for minor stations
    simplified = []
    for st in stations:
        if st in major_stations:
            simplified.append(st)
        else:
            simplified.append(f"Cluster_{st[0]}")  # Cluster by first letter

    # Add edges with train color
    for i in range(len(simplified)-1):
        u, v = simplified[i], simplified[i+1]
        if u != v:
            # If edge exists, keep the first color or leave as is
            if G.has_edge(u, v):
                continue
            G.add_edge(u, v, color=train_color)

print(f"Simplified graph: {len(G.nodes())} nodes, {len(G.edges())} edges")

# -------------------------------
# 5. Create PyVis Graph with Colors
# -------------------------------
net = Network(
    height="900px", 
    width="100%", 
    bgcolor="#111111",  # Dark background
    font_color="white", 
    notebook=False
)
net.force_atlas_2based(gravity=-50)

# Add nodes with size and color based on importance
for node in G.nodes():
    if node.startswith("Cluster_"):
        color = "#FFA500"  # Orange for clusters
        size = 5
    elif node in hub_stations:
        color = "#FF4500"  # Red for top hubs
        size = 20
    else:
        color = "#00BFFF"  # Blue for major stations
        size = 10
    
    net.add_node(node, label=node, color=color, size=size)

# Add edges with stored colors
for u, v, data_dict in G.edges(data=True):
    net.add_edge(u, v, color=data_dict.get('color', '#888888'))

# -------------------------------
# 6. Save Interactive Graph
# -------------------------------
output_file = "train_network_colored2.html"
net.save_graph(output_file)
print(f"Colored interactive graph saved as {output_file}. Open it in your browser.")
```
